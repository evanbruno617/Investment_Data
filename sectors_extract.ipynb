```python
import pandas as pd
import numpy as np
import os
import yfinance as yf
```

```python
#obtain list of sectors
directory = "sectors_list"
list_sector = []
for filename in os.listdir(directory):
    list_sector.append(filename)
```

```python
for sector in list_sector:

    symbols = pd.read_csv(f'sectors_list/{sector}')['0'].tolist()
    
    for i in symbols:
        
        
        
        print(i)
        # Set the start and end date
        start_date = '1990-01-01'
        end_date = '2023-01-01'
        
        try:

            # Get the data
            data = yf.download(i, start_date, end_date, interval='1mo')
            
        except:
            continue

        #Convert to dataframe
        data_df = pd.DataFrame(data)
    
        mypath = f"Sectors_Anlalysis/Monthly/{sector}"
        if not os.path.isdir(mypath):
            os.makedirs(mypath)

        upload_file = f"{mypath}/{i}.csv"
        data_df.to_csv(upload_file)
```
